```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('IceCreamData.csv')
df
```

```python
df['Temperature'] = (df['Temperature'] - 32) * 5 / 9
df
```

```python
df_sorted = df.sort_values('Temperature', ascending=False)
df_sorted
```

```python
plt.plot(df_sorted['Temperature'], df_sorted['Sales'], 'r.--', label='US$')
plt.title('Ice Cream Truck')
plt.ylabel('Sales (US$)')
plt.xlabel('Temperature (Celsius)')
plt.legend()

plt.savefig('icecream_result.pdf')
plt.show()
```
